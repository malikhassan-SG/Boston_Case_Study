```python
# Task 3: Use the appropriate tests to answer the questions provided.
# 3.1 Is there a significant difference in median value of houses bounded by the Charles river or not? (T-test for independent samples)
```

```python
Hypotheses:
Null hypothesis (H0): There is no significant difference in the median value of houses bounded by the Charles river and those not bounded by it.
Alternative hypothesis (H1): There is a significant difference in the median value of houses bounded by the Charles river and those not bounded by it.
Test: Perform an independent t-test between the two groups (bounded by Charles river vs not).
```

```python
# This code snippet will create a new column in the boston_df DataFrame called CHAS_T
boston_df.loc[(boston_df['CHAS'] == 0), 'CHAS_T'] = 'FAR'
boston_df.loc[(boston_df['CHAS'] == 1), 'CHAS_T'] = 'NEAR'
# This line of code will display the first 5 rows of the boston_df DataFrame.
boston_df.head(5)
```

```python
# T-test
# This line of code performs an independent two-sample t-test to compare the median values of owner-occupied homes (MEDV) for properties
# that are far from the Charles River (CHAS_T == 'FAR') and those that are near the Charles River (CHAS_T == 'NEAR')
t_statistic, p_value = scipy.stats.ttest_ind(boston_df[boston_df['CHAS_T'] == 'FAR']['MEDV'], 
                      boston_df[boston_df['CHAS_T'] == 'NEAR']['MEDV'], equal_var = True)
print(f"t_statistic: {t_statistic}\np_value: {p_value}")
```

```python
Given the p-value is less than 0.05, we reject the Null Hypothesis, meaning there is not a statistical difference in median value betwenn houses near the Charles River and houses far away.
```

```python
# 3.2 Is there a difference in Median values of houses (MEDV) for each proportion of owner occupied units built prior to 1940 (AGE)?
Hypotheses:
Null hypothesis (H0): There isn't statistical difference in Median values of houses (MEDV) for each proportion of owner occpied units built prior to 1940.
Alternative hypothesis (H1): There is statistical difference in Median values of houses (MEDV) for each proportion of owner occpied units built prior to 1940
Test: Perform ANOVA to compare the means of MEDV across the groups.
```

```python
# This code snippet uses the Statsmodels library to perform an Ordinary Least Squares (OLS) regression and 
# then conducts an Analysis of Variance (ANOVA) on the regression model.
# This line imports the ols function from the statsmodels.formula.api module
from statsmodels.formula.api import ols

# This line creates an OLS regression model
lm = ols('MEDV ~ AGE', data = boston_df).fit()

# This line performs an ANOVA on the fitted regression model using the anova_lm function from the 
# sm.stats module (which refers to Statsmodels). ANOVA is used to analyze the differences among 
# group means and their associated procedures.
table = sm.stats.anova_lm(lm)

# This line prints the ANOVA table.
print(table)
# df: Degrees of freedom for each source of variance
# sum_sq: Sum of squares for each source of variance.
# mean_sq: Mean sum of squares for each source of variance (sum_sq divided by df).
# F-statistic for the test.
# •	PR(>F): p-value for the F-test.
```

```python
Given p-value is less than 0.05, we fail to accept the Null Hypothesis --> There is statistical difference in Median values of houses (MEDV) 
for each proportion of owner occpied units built prior to 1940.
```

```python
# 3.2 - Can we conclude that there is no relationship between Nitric oxide concentrations and proportion of non-retail business acres per town? (Pearson Correlation)
Hypotheses:
Null hypothesis (H0): There is no relationship between Nitric oxide concentrations and the proportion of non-retail business acres.
Alternative hypothesis (H1): There is a relationship between Nitric oxide concentrations and the proportion of non-retail business acres.
Test: Perform Pearson correlation test.
```

```python
# The pearsonr function computes the Pearson correlation coefficient, which measures the linear correlation between two variables. It returns two values:
# correlation coefficient: A value between -1 and 1 that indicates the strength and direction of the linear relationship between the two variables.
# A value close to 1 indicates a strong positive correlation.
# A value close to -1 indicates a strong negative correlation.
# A value close to 0 indicates no linear correlation.
# p-value: The probability of observing the data given that there is no linear correlation between the variables (the null hypothesis). 
# A small p-value (typically < 0.05) indicates strong evidence against the null hypothesis.

# This line of code uses the pearsonr function from the scipy.stats module to calculate the Pearson correlation 
correlation, p_value = scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
print(f"t_statistic: {t_statistic}\np_value: {p_value}")
```

```python
Given the Pearson Coefficient is 0.76365 and p-value less than 0.05, we reject the Null Hypothesis as there is a positive correlation between 
Nitric oxide concentration and proportion of non-retail business acres per town.
```

```python
#3.4 What is the impact of an additional weighted distance to the five Boston employment centres on the median value of owner occupied homes? (Regression analysis)
Hypotheses:
Null hypothesis (H0): The weighted distances to employment centres have no impact on the median value of homes.
Alternative hypothesis (H1): The weighted distances to employment centres have an impact on the median value of homes.
Test: Perform a linear regression analysis where MEDV is the dependent variable and the weighted distance to employment centres is the independent variable.
```

```python
# This code snippet performs an Ordinary Least Squares (OLS) regression to model the relationship between the distance to employment centers (DIS) 
# and the median value of owner-occupied homes (MEDV).
# This code extracts the DIS column (distance to employment centers) as the independent variable x 
# and the MEDV column (median value of owner-occupied homes) as the dependent variable y.
x = boston_df['DIS']
y = boston_df['MEDV']

# This line of code adds a constant term (intercept) to the independent variable 
x = sm.add_constant(x)

# This line of code creates and fits an OLS regression model with y as the dependent variable and x 
model = sm.OLS(y, x).fit()

# This line of code generates predictions based on the fitted model.
predisction = model.predict(x)

# This line of code prints a comprehensive summary of the regression results
model.summary()
```
